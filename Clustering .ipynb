# K Means Clustering



* K Means Clustering is an unsupervised learning algorithm 


* The K is K Means is called as a centroid  

__Steps :__
1. Inititlize the centroid or k value randomly (Eg. K=2)
2. Find the distance between the data points and the centroids .
3. The distance is calculated with the help of Euclidien distance or Manhatten distance .
4. With the help of previous step we divide our data into groups as per the no. of centroids .
5. find the mean value of each group and move the centroid to the center position .
6. Repeat the step and 5 until we have fixed no. of groupes with centroids in the center of every group .

# DBSCAN 

* DBSCAN stands for Density Based Spatial Clustering of Applications with Noise .

* It is one of the unsupervised ml algo which some times perform better then KMeans and Hierarchecal Clustering .









__Practical :__



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

data = 'https://raw.githubusercontent.com/krishnaik06/DBSCAN-Algorithm/master/Mall_Customers.csv'

df = pd.read_csv(data)

# Independent Features
X = df.iloc[:,[3,4]].values

# Apply db scan from sklearn
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=3,min_samples=4)

# Fit the model
model = dbscan.fit(X)

labels = model.labels_
labels

# Here -1 is noisy points and the clusters are in the range of 0 to 1
```

```python
# Identify the points which make up our core points

sample_cores = np.zeros_like(labels,dtype=bool)
sample_cores[dbscan.core_sample_indices_] = True
sample_cores

# In this the points wich are -1 are represented as False and the points
# othre then -1  are represented as True
```

```python
# Calculating the number of clusters

n_clusters = len(set(labels))-(1 if -1 in labels else 0)

# Evaluate 

from sklearn import metrics

print(metrics.silhouette_score(X,labels))
```

__Practical :__

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples,silhouette_score
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
```

```python
# Generating the sample data from make_blobs
# This particular setting has one distinct cluster and 3 clusters placed close
# together.
X, y = make_blobs(n_samples=500,
                  n_features=2,
                  centers=4,
                  cluster_std=1,
                  center_box=(-10.0, 10.0),
                  shuffle=True,
                  random_state=1)  # For reproducibility
```

__Elbow Method :__

```python
wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('No. of Cluster')
plt.ylabel('WCSS')
plt.show()

clus = KMeans(n_clusters=4,random_state=10)
clus_labels = clus.fit_predict(X)
print(clus_labels)
```

__Entropy :__



* In order to consider which feature to select for splitting the node .
* It defines the randomness in the data
* Entropy is a metric which will measures the purity of the split .
* It is the first step to solve the problem of a decision tree

* Entropy value ranges between 0 and 1 .
* If 1 then completely impure subset and if 0 it is completely pure subset .
* If we get the value as 0 we will stop further splitting and consider it as leaf node .
* In entropy we will just calculate whether the split is pure or not .
* Entropy is used only for one node if we need to calculate for all the nodes till the depth we will use information gain .

__Information gain :__

* Here we compute the average of all the entropy

__Design a Decision tree :__

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
iris = load_iris()
X = iris.data[:,2:]
y = iris.target
tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(X,y)
```

## WHICH MODEL TO CHOOSE FROM :-

It depend on the type of problem we are solving :
> * If the relationship b/w dependent and independet variable is well approximated by linear model then linear regression will outperform tree based model .
> * If there is high nonlinearity and complex relationship b/w dependent and independent var a tree model will outperform a classification regression problem .
> * If we need a model which is easy to explain to people the tree model will do the job for us .

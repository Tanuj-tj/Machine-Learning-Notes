__Confusion Matrics :__

* FP (False Positive) or Type 1 error
* FP (False Negetive) or Type 2 error
* We need to reduce the type 1 and type 2 error or FP and FN
* If we have a balance dataset (500 points belongs 1 , 500 points belongs to 0) we compute the accuracy .

__Accuracy :__


* If we have a imblanced dataset like 900 points belongs to 1 and 100 points belongs to 0 , If we compute accuracy then out model compute all points as class 1 and we will get a very high accuracy .

* If we have imblanced dataset then we use recall or precission .



__Recall :__



* Out of the total positive values how many of them did we predicted correctly .

* It is also known as true positive rate and sensitivity .

__Precision :__



* Out of the totla predicted positive results how many results are actually positive .
* Also known as positive predicted value .

> Eg. 
> 1. Whenever the false positive is much more importat then we have to use precision .Like in case of spam mail detection if the mail is not spam and the model predicted it as a spam the person might miss an important mail .

> 2. Whenever the false negetive is important we have to use recall , like in the case of detecting whether a person is having cancer or not if a person is having cancer and we detet that the person does not have cancer then it might be very dangerous .

* In some cases both precision and recall are important in that case we should use F Beta . 

__F Beta score :__

> The above formula of f1 score is similar to Harmonic mean if we consider precision as x and recall as y

* When FP and FN both are equally important then we will considet the beta value as 1
* When FP is having more impact then the FN then we will reduce the beta value to 0.5 or between 0 to 1 .
* When FN is having more impact then the FP then we will increase the beta value to 2 .

__ROC curve and AUG__

* Mostly used for binary classification problems
* ROC (Receiver Operating characterisitcs curve)
* AUC (Area Under the ROC curve)



* An ROC curve plots TPR vs. FPR at different classification thresholds. Lowering the classification threshold classifies more items as positive, thus increasing both False Positives and True Positives. The following figure shows a typical ROC curve.



* To compute the points in an ROC curve, we could evaluate a logistic regression model many times with different classification thresholds, but this would be inefficient. Fortunately, there's an efficient, sorting-based algorithm that can provide this information for us, called AUC.



* The Area Under the Curve (AUC) is the measure of the ability of a classifier to distinguish between classes and is used as a summary of the ROC curve.

* The higher the AUC, the better the performance of the model at distinguishing between the positive and negative classes.

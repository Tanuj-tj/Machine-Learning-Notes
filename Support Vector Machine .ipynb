# SVM 

Link : https://towardsdatascience.com/support-vector-machine-introduction-to-machine-learning-algorithms-934a444fca47

* It is a supervised learning algorithm
* It is used to solve both classification and regression problems 
* Support vector machine is highly preferred by many as it produces significant accuracy with less computation power.

It consists of :
1. Support Vectors
2. Hyperplanes
3. Marginal Distances
4. Linear Seperable points
5. Non Linear Seperable Points



### SVM Kernels 

* It is used to add more dimention to a lower dimention space to make to easier to segrigate the data .
* It converge the inseperable problem to a seperable problem by adding more dimentions using the kernal trick  .
* The kernal trick help to make a more accurate classifier .

__Type of Kernals :__

1. Linear Kernals 
> Used as a normal dot product between any two given observatons .
2. Polymonial Kernals
> It is a generalised form of Linear Kernal . It can distinguish curves and non linear input space as well .
3. Radial Basis Function Kernel 
> It can map the space in infinite dimentions .
